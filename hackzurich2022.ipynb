## personal statistics

```python
import numpy as np

def get_personal_statistics(runner_data):
  statisticts = dict()
  statisticts['average_speed_2d'] = np.average(runner_data['speed_2D[km/h]'])
  statisticts['average_speed_3d'] = np.average(runner_data['speed_3D[km/h]'])

  start = runner_data['time[ISO-UTC]'].to_list()[0].split('T')[1]
  finish = runner_data['time[ISO-UTC]'].to_list()[-1].split('T')[1]

  duration = int(finish) - int(start)
  statisticts['duration'] = duration/100000 #in minutes
  statisticts['max_speed_2d'] = np.max(runner_data['speed_2D[km/h]'])
  statisticts['max_speed_3d'] = np.max(runner_data['speed_3D[km/h]'])
  statisticts['alt_max'] = np.max(runner_data['alt_baro[m]']) #TO_DO ask what is the diff
  statisticts['alt_min'] = np.min(runner_data['alt_baro[m]']) #TO_DO ask what is the diff
  
  return statisticts

```

```python
for runner in runners:
  statisticts = get_personal_statistics(runner)
  print(statisticts)
```
